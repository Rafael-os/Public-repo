```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/Kaunas/visit.kaunas.lt/master/Nurinkta%20%C4%AF%20CSV/to-see.csv"
csv = pd.read_csv(url)  #using panda to retrieve information from url in csv format
csv
```

```python
csv.info()  #show overall statistics of csv
```

```python
# retrieve in a list all different categories
csv_category = csv.drop_duplicates(subset = "Category")
csv_category["Category"].tolist()
```

```python
# Display how many rows are in each category
amount = pd.DataFrame(csv, columns = ['Category'])
amount.pivot_table(index = ["Category"], aggfunc = 'size').sort_values(ascending = False)
```
